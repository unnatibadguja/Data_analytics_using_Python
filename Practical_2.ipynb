```python
def generate_fibonacci_sequence(limit):
    fib_sequence = [0,1]
    while fib_sequence[-1] + fib_sequence[-2] <= limit:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
        print("Febonacci Sequence:", fib_sequence)
        
user_limit= int(input("Enter a limit for the Fibonacci Sequence:"))
generate_fibonacci_sequence(user_limit)
        
```

```python
dict1={"Pranjal":21,"Aparna":34}
dict2={"Vaishali":21,"Aparna":12}
dict_combined=dict1
for key,value in dict2.items():
    if key in dict_combined:
        dict_combined[key] +=value
    else:
        dict_combined[key] =value
dict_combined                
```
