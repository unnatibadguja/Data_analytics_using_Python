**3. Quadratic Equation Solver**  
Write a Python script that solves a quadratic equation of the form (ax^2 + bx + c = 0) given user input for coefficients (a), (b), and (c).

```python
import numpy as np
a=float(input("A:"))
b=float(input("B:"))
c=float(input("C:"))
x1=(-b+np.sqrt(np.pow(b,2)-4*a*c))/(2*a)
x2=(-b-np.sqrt(np.pow(b,2)-4*a*c))/(2*a)
print("Solution to the given quadratic equation {}x^2+{}x+{} are {} and {}".format(a,b,c,x1,x2))
        
```

## Q5

**5. Character Counter**  
Develop a Python script that takes a user input string and counts the number of occurrences of each character in the string.

```python
text=input("Enter a string:")
char_count={}
for char in text:
    if char in char_count:
        char_count[char] +=1
    else:
        char_count[char] =1
        print("\nCharecter Frequency:")
        for char, count in char_count.items():
            print(f"'{char}': {count}")
```

## Q6

**6. Develop a Python program that calculates the factorial of a given non-negative integer entered by the user. Display the result with a suitable message.**

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

user_integer = int(input("Enter a non-negative integer: "))
result = factorial(user_integer)
print(f"The factorial of {user_integer} is {result}.")
```
